```python
import pickle

file = open("Throughput_Small.pkl",'rb')
ds1 = pickle.load(file)
```

```python
ds1 = ds1.drop('nrStatus', axis=1)
ds1.head()
```

```python
ds1['Throughput'] = 0
ds1.head()
```

```python
ds1_selected_feature = ds1[ds1.columns[3:]]
ds1_selected_feature.head()
```

```python
file = open("Throughput_Medium.pkl",'rb')
ds2 = pickle.load(file)
ds2['Throughput'] = 1
ds2 = ds2.drop('nrStatus', axis=1)
ds2_selected_feature = ds2[ds2.columns[3:]]
ds2_selected_feature.head()
```

```python
file = open("Throughput_Large.pkl",'rb')
ds3 = pickle.load(file)
ds3['Throughput'] = 2
ds3 = ds3.drop('nrStatus', axis=1)
ds3_selected_feature = ds3[ds3.columns[3:]]
ds3_selected_feature.head()
```

```python
import pandas as pd

ds_combined = pd.concat([ds1_selected_feature,ds2_selected_feature,ds3_selected_feature])
ds_combined.head()
```

```python
df_shuffled = ds_combined.sample(frac=1).reset_index(drop=True)
df_shuffled.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

df = df_shuffled.copy()

# 'Throughput' is the target and all other columns are features
X = df.drop('Throughput', axis=1)
y = df['Throughput']
# Normalize the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42, shuffle=True,stratify=y)

# Normalize the feature set
# scaler = StandardScaler()
# X_train = scaler.fit_transform(X_train)
# X_test = scaler.transform(X_test)

```

```python
# Initialize models
models = {
    "Random Forest": RandomForestClassifier(n_estimators=200),
    "XGBoost": XGBClassifier(n_estimators = 200,use_label_encoder=False, eval_metric='mlogloss'),
    "MLP": MLPClassifier(hidden_layer_sizes=(100, 100,50 ), max_iter=300),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(n_neighbors=10)
}

# Train models
for name, model in models.items():
    model.fit(X_train, y_train)
    print(f"{name} trained.")
```

```python
# Evaluate models
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"Model: {name}")
    print(classification_report(y_test, y_pred))
    print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Predict using Random Forest
rf = models["Random Forest"]
y_pred_rf = rf.predict(X_test)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
plt.title('Confusion Matrix for Random Forest')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
