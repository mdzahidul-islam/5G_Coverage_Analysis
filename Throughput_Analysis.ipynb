```python
#Throughput Data
Throughput_df = pd.read_csv('/content/drive/MyDrive/THROUGHPUT DATA.csv')
Throughput_df.head(10)
```

```python
Throughput_df = Throughput_df[['run_num', 'seq_num', 'abstractSignalStr', 'latitude', 'longitude', 'movingSpeed', 'compassDirection', 'nrStatus', 'lte_rssi', 'lte_rsrp', 'lte_rsrq', 'lte_rssnr', 'nr_ssRsrp', 'nr_ssRsrq', 'Throughput']]
Throughput_df = Throughput_df.dropna()
Throughput_df.head(10)
```

```python
Throughput_small_df = Throughput_df[Throughput_df['Throughput'] < 300]
Throughput_medium_df = Throughput_df[(Throughput_df['Throughput'] >= 300) & (Throughput_df['Throughput'] <= 700)]
Throughput_large_df = Throughput_df[Throughput_df['Throughput'] > 700]

Throughput_small_df.head(10)
```
